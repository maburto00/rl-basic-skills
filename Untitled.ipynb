# Graphs for RL performance



```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

total_frames=2000
fig = plt.figure()
ax = fig.add_subplot() 
x=np.array([i for i in range(2000)])
y1=np.array([1*i for i in x])
y2=np.array([2*i for i in x])
y3=np.array([3*i for i in x])
plt.xlabel("Frames")
plt.ylabel("Human-normalized Score")
ax.plot(x,y1)
ax.plot(x,y2)
ax.plot(x,y3)
```
